```python
import GEOparse
import pandas as pd

# Download GEO dataset
gse = GEOparse.get_GEO(geo="GSE134379", destdir="./")

# Check available metadata
print("Dataset title:", gse.metadata.get("title", "Unknown"))
print("Number of samples:", len(gse.gsms))

# Extract expression or methylation data
expression_data = {}
for gsm_name, gsm in gse.gsms.items():
    # GSM table may contain the data (depends on the dataset)
    if not gsm.table.empty:
        expression_data[gsm_name] = gsm.table['VALUE']

# Convert expression data to a DataFrame
expression_df = pd.DataFrame(expression_data)
print("Expression data shape:", expression_df.shape)
```

```python

import pandas as pd
from bs4 import BeautifulSoup

# Load the HTML file
with open("GEO2R - GEO - NCBI.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML content
soup = BeautifulSoup(html_content, "html.parser")

# Find all tables in the HTML
tables = soup.find_all("table")

# If there are multiple tables, choose the appropriate one
# For simplicity, we'll use the first table
if tables:
    table_html = str(tables[0])  # Convert the table to a string
    # Use pandas to read the table
    df = pd.read_html(table_html)[0]

    # Show the extracted data
    # print(df)

    # Save the data to a CSV file if needed
    df.to_csv("extracted_data.csv", index=False)
else:
    print("No tables found in the HTML file.")


df
```
