```python
!pip install Bio

from Bio import Entrez
from Bio import SeqIO

Entrez.email = "rozhin.maleki@yahoo.com"  # Replace with your email address

def get_gene_sequence(gene_name, database="nucleotide"):
    search_term = gene_name + "[Gene Name]"
    handle = Entrez.esearch(db=database, term=search_term)
    record = Entrez.read(handle)
    gene_id = record["IdList"][0]
    handle = Entrez.efetch(db=database, id=gene_id, rettype="fasta", retmode="text")
    sequence = SeqIO.read(handle, "fasta")
    return sequence

gene_name = "BRCA2"
sequence = get_gene_sequence(gene_name)
print(sequence.seq)
```

# 2D kmers count matrix

```python
import numpy as np
import itertools
import pandas as pd



def count_kmers(sequence, ksize):
    count_matrix = {}
    n_kmers = len(sequence) - ksize + 1

    # Initialize count_matrix with all possible k-mers
    all_kmers = [''.join(kmer) for kmer in itertools.product('ACGT', repeat=ksize)]
    for kmer in all_kmers:
        count_matrix[kmer] = 0

    # Count occurrences of k-mers in the sequence
    for i in range(n_kmers):
        kmer = sequence[i:i + ksize]
        count_matrix[kmer] += 1

    return count_matrix

sequence = 'ATGGACCAGATATAGGGAGAGCCAGGTAGGACAAAATTATTACGTCTAATCATCATCATCAGAGGATCATCA'
ksize = 3
count_matrix = count_kmers(sequence, ksize)

print(count_matrix)



# sequence = 'ATGGACCAGATATAGGGAGAGCCAGGTAGGACAAAATTATTACGTCTAATCATCATCATCAGAGGATCATCA'
# ksize = 8
# count_matrix = count_kmers(sequence, ksize)

# Convert count_matrix to a matrix shape of 256x256
matrix = np.zeros((2**ksize *2**ksize, 2**ksize *2**ksize))

counts = []
patterns = []
for i, kmer in enumerate(count_matrix):
    # x = ord(kmer[0])
    # y = ord(kmer[1])
    # matrix[x][y] = count_matrix[kmer]

    counts.append(count_matrix[kmer])
    patterns.append(kmer)

    # print(i,x,y ,'x o y were announced',count_matrix[kmer])

# print(matrix)
```
